```python
from obspy.taup import TauPyModel
from obspy.geodetics import degrees2kilometers
import matplotlib.pyplot as plt
from obspy.taup import plot_travel_times

import numpy as np


#########################CLASS def###################

class SisEvnt:
    """Almacena informacion relevante de un evento"""

    def __init__(self, station, Tsp):
        self.name = station
        self.deltaT = Tsp
        self.d = 0
        self.up = 0
        self.dw = 0
        self.flag = False

#####################################################
fig, ax = plt.subplots(figsize=(9, 9))

#ax = plot_travel_times(source_depth=10, phase_list=["P", "S"], ax=ax, fig=fig, verbose=True)

#Km = degrees2kilometers(30) #30 grados a km

model = TauPyModel(model="iasp91")

#arrivals = model.get_travel_times(source_depth_in_km=55, distance_in_degree=30, phase_list=["P", "S"])

#ArTime = arrivals[0]

#Vp = Km/arrivals[0].time
#Vs = Km/arrivals[1].time
####### Delta Ts-p de cada estacion ###################
Tsdv = 90.21606
Tgr  = 27.91332
Tbin = 47.457
Tsam = 51.43418

St1 = SisEvnt('SDV',Tsdv)
St2 = SisEvnt('GR',Tgr)
St3 = SisEvnt('BIN',Tbin)
St4 = SisEvnt('SAM',Tsam)
#####################################################
deep = 34 # en Km

Ptimes=[]
Stimes=[]
SPtimes=[]
dist=[]

for n in range(50):
    
    Ptimes.append(n)
    Stimes.append(n)
    SPtimes.append(n)
    dist.append(n)
    ArrTimesP = model.get_travel_times(source_depth_in_km=deep, distance_in_degree=(n+1)*0.5, phase_list=["P"])
    ArrTimesS = model.get_travel_times(source_depth_in_km=deep, distance_in_degree=(n+1)*0.5, phase_list=["S"])
    #print(n)
   # print(ArrTimesP)
   # print(ArrTimesS)
    if len(ArrTimesP)!= 0 :
        Ptimes[n]=ArrTimesP[0].time
        Stimes[n]=ArrTimesS[0].time
        SPtimes[n]=Stimes[n]-Ptimes[n]
    dist[n]=(n+1)*0.5
    
    if St1.deltaT <= SPtimes[n] and St1.flag == False :
        St1.up = (n+1)*0.5
        St1.dw = (n)*0.5
        St1.flag = True
        print("sdv cota ",SPtimes[n-1],"-", SPtimes[n])
    if St2.deltaT <= SPtimes[n] and St2.flag == False :
        St2.up = (n+1)*0.5
        St2.dw = (n)*0.5
        St2.flag = True
        print("gr cota ",SPtimes[n-1],"-", SPtimes[n])
    if St3.deltaT <= SPtimes[n] and St3.flag == False :
        St3.up = (n+1)*0.5
        St3.dw = (n)*0.5
        St3.flag = True
        print("bin cota ",SPtimes[n-1],"-", SPtimes[n])
    if St4.deltaT <= SPtimes[n] and St4.flag == False :
        St4.up = (n+1)*0.5
        St4.dw = (n)*0.5
        St4.flag = True
        print("sam cota ",SPtimes[n-1],"-", SPtimes[n])
        

for n in np.arange(St1.dw, St1.up+0.01, 0.01):
    ArrTimesP = model.get_travel_times(source_depth_in_km=deep, distance_in_degree=n, phase_list=["P"])
    ArrTimesS = model.get_travel_times(source_depth_in_km=deep, distance_in_degree=n, phase_list=["S"])
    #print(n)
    #print(ArrTimesS[0].time - ArrTimesP[0].time)
    if St1.deltaT <= (ArrTimesS[0].time - ArrTimesP[0].time) :
        St1.up = n
        St1.dw = n-0.1
        St1.d = degrees2kilometers(n-0.05)
        print("Distancia de ", St1.name,": ", St1.d, "Km")
        break
    
for n in np.arange(St2.dw, St2.up+0.01, 0.01):
    ArrTimesP = model.get_travel_times(source_depth_in_km=deep, distance_in_degree=n, phase_list=["P"])
    ArrTimesS = model.get_travel_times(source_depth_in_km=deep, distance_in_degree=n, phase_list=["S"])
    #print(n)
    #print(ArrTimesS[0].time - ArrTimesP[0].time)
    if St2.deltaT <= (ArrTimesS[0].time - ArrTimesP[0].time) :
        St2.up = n
        St2.dw = n-0.1
        St2.d = degrees2kilometers(n-0.05)
        print("Distancia de ", St2.name,": ", St2.d, "Km")
        break
        
for n in np.arange(St3.dw, St3.up+0.01, 0.01):
    ArrTimesP = model.get_travel_times(source_depth_in_km=deep, distance_in_degree=n, phase_list=["P"])
    ArrTimesS = model.get_travel_times(source_depth_in_km=deep, distance_in_degree=n, phase_list=["S"])
    #print(n)
    #print(ArrTimesS[0].time - ArrTimesP[0].time)
    if St3.deltaT <= (ArrTimesS[0].time - ArrTimesP[0].time) :
        St3.up = n
        St3.dw = n-0.1
        St3.d = degrees2kilometers(n-0.05)
        print("Distancia de ", St3.name,": ", St3.d, "Km")
        break
for n in np.arange(St4.dw, St4.up+0.01, 0.01):
    ArrTimesP = model.get_travel_times(source_depth_in_km=deep, distance_in_degree=n, phase_list=["P"])
    ArrTimesS = model.get_travel_times(source_depth_in_km=deep, distance_in_degree=n, phase_list=["S"])
    #print(n)
    #print(ArrTimesS[0].time - ArrTimesP[0].time)
    if St4.deltaT <= (ArrTimesS[0].time - ArrTimesP[0].time) :
        St4.up = n
        St4.dw = n-0.1
        St4.d = degrees2kilometers(n-0.05)
        print("Distancia de ", St4.name,": ", St4.d, "Km")
        break
        

ax.plot(dist,Ptimes)
ax.plot(dist,Stimes)
ax.plot(dist,SPtimes)
plt.show()
####################################################
print(" ")
print("Tiempos P")
print(Ptimes)
print(" ")
print("Tiempos P")
print(" ")
print(Stimes)
print(" ")
print("Tiempos S-P")
print(SPtimes)
```
